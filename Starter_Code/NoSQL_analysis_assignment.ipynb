## Notebook Set Up

```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question: 
* Use `count_documents` to display the number of documents contained in the result.
* Display the first document in the results using `pprint`.

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
hygiene = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count = db.establishments.count_documents(hygiene)
pprint(count)
pprint("")
# Display the first document in the results using pprint
h_list = list(establishments.find(hygiene))
pprint(h_list[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.json_normalize(h_list)
# Display the number of rows in the DataFrame
print(len(hygiene_df))
# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4. LOCAL AUTHORITY IS NOT LONDON, ITS CITY OF LONDON COPORATION
London = {"AddressLine2": 'London', "RatingValue": {'$gte': 4}}

# Use count_documents to display the number of documents in the result
count = db.establishments.count_documents(London)
pprint(count)
pprint("")
# Display the first document in the results using pprint
lon_list = list(establishments.find(London))
pprint(lon_list[0])
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.json_normalize(lon_list)
# Display the number of rows in the DataFrame
print(len(london_df))
# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.086058
longitude = 1.196408
rating_value = 5

query = {'geocode.latitude': {'$gte': latitude-degree_search, '$lt': latitude+degree_search},
         'geocode.longitude': {'$gte': longitude-degree_search, '$lt': longitude+degree_search},
         'RatingValue': rating_value}
# filter = {'BusinessName': 1}
sort = [('scores.Hygiene', -1)]
limit = 5

# Print the results
lat_list = list(establishments.find(query).sort(sort).limit(limit))
pprint(lat_list)
```

```python
# Convert result to Pandas DataFrame
lat_df = pd.json_normalize(lat_list)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': {"LocalAuthorityName": "$LocalAuthorityName"},'count': { '$sum': 1 }}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': { 'count': -1 }}

# Print the number of documents in the result
pipeline = [match_query, group_query, sort_values]

# Print the first 10 results
results = list(establishments.aggregate(pipeline))
print("Number of rows in result: ", len(results))
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.json_normalize(results)
# Display the number of rows in the DataFrame
print(len(results_df))
# Display the first 10 rows of the DataFrame
results_df.head(10)
```
